# creating a chatbot using speech recogniation library

```python
import speech_recognition as sr
import random
from random import choice,shuffle
r=sr.Recognizer()
def voice():
    with sr.Microphone() as source:
        print("बोल्नुहोस् !");
        audio=r.listen(source)
        global text
        try:
            
            text =r.recognize_google(audio)
        
        except:
            text="माफ गर्नुहोस् हामी तपाईंको आवाज बुझ्न असमर्थ छौं |\n"
def user():
    voice()
    print("तपाईंले {} भन्नुभयो".format(text))
    
def main():
    user()
    while True:
        words=["namaste","hello","hi",]
        
        if ((text.lower()) in words):
            print(choice(["नमस्ते!","नमस्ते!,स्वागत छ हजुरलाई",]))
            user()
        if (text in ["tapai ko naam ki ho","naam","name","Naam"]):
            print("मेरो नाम पाइथन च्याटबट हो |")
            print("तपाईंको नाम के हो?")
            user()
            print("तपाईंलाई धन्यवाद |")
        else:
            user()
if __name__ == "__main__":
    main()
```
