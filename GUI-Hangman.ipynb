```python
import random
import tkinter as tk
from tkinter import messagebox

global chosen_word, player_guess, guessed_letters, word_guessed, joined_word, attempts, words, picture, new_game

new_game = True

words = ["Hello", "Python", "Workshop", "Function", "Variable", "Programming", "Prize", "Loop", "Output", "Input",
         "String", "Reverse", "Palindrome", "Structures", "","Dictionary", "Exception", "Inheritance", 
         "Object", "Precedence"]
picture = (
"""
-----
|      |
|      
|  
|       
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
| 
|       
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
|    -+- 
|       
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
|   /-+- 
|       
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
|  /-+-\ 
|       
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
|  /-+-\ 
|      | 
|       
|
|
|
--------
""",
"""
-----
|      |
|      0
|  /-+-\ 
|      | 
|      | 
|
|
|
--------
"""
)

root = tk.Tk()
root.title("Hangman")

info_frame = tk.Frame(root)
info_frame.grid(row=0,column=0)

dotted_font = "Arial 20 bold"

attempts_info_val = tk.StringVar()
guess_show_val = tk.StringVar()
status_val = tk.StringVar()
end_val = tk.StringVar()

welcome = tk.Label(info_frame, text="Welcome to Hangman Game!")
welcome.grid(row=0,column=0)

game_frame = tk.Frame(root)
game_frame.grid(row=1,column=0)

hang_pic = tk.Label(game_frame, text="",justify=tk.LEFT)
hang_pic.grid(row=0,column=0)

entry_frame = tk.Frame(game_frame)
entry_frame.grid(row=0,column=1)

attempts_info = tk.Label(entry_frame, textvariable=attempts_info_val)
attempts_info.grid(row=0,column=0,columnspan=2)

guess_show = tk.Label(entry_frame, textvariable=guess_show_val,font=dotted_font)
guess_show.grid(row=1,column=0,columnspan=2)

prompt = tk.Label(entry_frame, text="Please enter a letter between A-Z and Click on guess")
prompt.grid(row=2,column=0,columnspan=2)

guess = tk.Entry(entry_frame)
guess.grid(row=3,column=0)


status_frame = tk.Frame(root)
status_frame.grid(row=2,column=0)

status = tk.Label(status_frame, textvariable=status_val)
status.grid(row=0,column=0)

end = tk.Label(status_frame, textvariable=end_val)
end.grid(row=1,column=0)

newgm_frame = tk.Frame(root)
newgm_frame.grid(row=3,column=0)

def toggle_exitbtn():
    play = tk.Button(newgm_frame, text='New Game', command=main)
    play.grid(row=1,column=0)
    exit = tk.Button(newgm_frame, text='Exit', command=root.destroy)
    exit.grid(row=1,column=1)

def reset():
    global chosen_word, player_guess, guessed_letters, word_guessed, joined_word, attempts, words, picture, new_game
    chosen_word = random.choice(words).lower()
    player_guess = None # for hold the players guess
    guessed_letters = [] # list of guessed letters
    word_guessed = []
    for letter in chosen_word:
        word_guessed.append("-") # create an unguessed, blank version of the word
    joined_word = None # joins the words in the list word_guessed
    attempts = len(picture) - 1
    print(chosen_word)

def update():
    end_val.set(" ")
   
    status_val.set(" ")
    
    global chosen_word, player_guess, guessed_letters, word_guessed, joined_word, attempts, words, picture, new_game
    
    hang_pic = tk.Label(game_frame, text=picture[(len(picture) - 1) - attempts],justify=tk.LEFT)
    hang_pic.grid(row=0,column=0)
    
    
    attempts_info_val.set("Attempts Left: "+str(attempts))
    
    
    if new_game == True:
        joined_word = "".join(word_guessed)
        guess_show_val.set(joined_word)
    
    
    if new_game == False:
        print("Game is new_game = false")
        try:
            player_guess = str(guess.get()).lower()
            print("player guess: ", player_guess)
        except: # check valid input
            print("That is not valid input. Please try again.")           
            status_val.set("That is not valid input. Please try again.")
        else: 
            if not player_guess.isalpha(): # checks if the input is a letter. Also checks an input has been made.
                print("That is not a letter. Please try again.")
                status_val.set("That is not a letter. Please try again.")
            elif len(player_guess) > 1: # checks if  the input is only one letter
                print("That is more than one letter. Please try again.")
                status_val.set("That is more than one letter. Please try again.")
            elif player_guess in guessed_letters: # check it letter hasn't been guessed already
                print("You have already guessed that letter. Please try again.")
                status_val.set("You have already guessed that letter. Please try again.")
            else:
                pass
            
        guessed_letters.append(player_guess)

        for letter in range(len(chosen_word)):
            if player_guess == chosen_word[letter]:
                word_guessed[letter] = player_guess # replaces all letters in the chosen word that match the players guess
                joined_word = "".join(word_guessed)
                guess_show_val.set(joined_word)
    
        if player_guess not in chosen_word:
            attempts = attempts-1
            attempts_info_val.set("Attempts Left: "+str(attempts))
            hang_pic = tk.Label(game_frame, text=picture[(len(picture) - 1) - attempts],justify=tk.LEFT)
            hang_pic.grid(row=0,column=0)

        

        if "-" not in word_guessed: # no blanks remaining
            status_val.set("\nCongratulations! you guessed the word "+chosen_word)
            toggle_exitbtn()
        elif attempts==0: # loop must have ended because attempts reached 0
            status_val.set("\nSorry! The word was "+chosen_word)
            toggle_exitbtn()
            
            
        guess.delete(0,tk.END)
    
guessbtn = tk.Button(entry_frame, text="Guess", command=update)
guessbtn.grid(row=3,column=1)

def main():
    global new_game
    
    new_game = True
    
    if new_game==True:
        reset()
        update()
        new_game = False

main()
root.mainloop()
```
